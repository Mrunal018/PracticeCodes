```python
class FinEdRegistration:
    def __init__(self):
        self.registered_users = []

    def explain_benefits(self):
        print("\n--- Why Register for FinEdu? ---")
        print("Before you proceed with registration, here’s why joining FinEdu is a great decision:")
        print("1. **Personalized Learning:** Access tailored financial education resources designed to match your goals.")
        print("2. **Interactive Tools:** Use practical tools like budgeting calculators and investment trackers to manage your finances effectively.")
        print("3. **Gamified Experience:** Engage with fun quizzes and challenges that make learning about finance enjoyable.")
        print("4. **Supportive Community:** Connect with a network of peers and experts who share valuable financial insights.")
        print("5. **Career Enhancement:** Improve your financial literacy—a skill highly valued by employers and crucial for personal success.")
        print("Let's get you started on this journey!")

    def register_user(self):
        print("\n--- FinEdu Registration ---")
        print("To register, please provide the following details.")

        while True:
            name = input("Enter your name: ").strip()
            if not name:
                print("Name cannot be empty. Please try again.")
                continue

            age_input = input("Enter your age: ").strip()
            try:
                age = int(age_input)
            except ValueError:
                print("Invalid age entered. Please enter a numeric value.")
                continue

            if 18 <= age <= 30:
                username = input("Create a username: ").strip()
                password = input("Create a password: ").strip()

                self.registered_users.append({
                    "name": name,
                    "age": age,
                    "username": username,
                    "password": password
                })
                print(f"Thank you, {name}! You have successfully registered for FinEdu.")
                break
            else:
                print("Sorry, registration is only open to individuals between the ages of 18 and 30. Please try again.")

    def login_user(self):
        print("\n--- FinEdu Login ---")
        username = input("Enter your username: ").strip()
        password = input("Enter your password: ").strip()

        for user in self.registered_users:
            if user['username'] == username and user['password'] == password:
                print(f"Welcome back, {user['name']}! You have successfully logged in to FinEdu.")
                return
        print("Login failed. Please check your username and password, and try again.")

    def display_registered_users(self):
        if self.registered_users:
            print("\n--- Registered Users ---")
            for user in self.registered_users:
                print(f"- {user['name']}, Age: {user['age']}, Username: {user['username']}")
        else:
            print("No users have registered yet.")

    def run(self):
        self.explain_benefits()

        while True:
            print("\n--- FinEdu Main Menu ---")
            print("1. Register User")
            print("2. Login")
            print("3. Display Registered Users")
            print("4. Exit")
            choice = input("Enter your choice (1/2/3/4): ").strip()

            if choice == '1':
                self.register_user()
            elif choice == '2':
                self.login_user()
            elif choice == '3':
                self.display_registered_users()
            elif choice == '4':
                print("Thank you for using FinEdu. Goodbye!")
                break
            else:
                print("Invalid choice. Please select a valid option.")

# Create an instance of the FinEdRegistration class
fin_ed_registration = FinEdRegistration()

# Run the application
fin_ed_registration.run()
```
